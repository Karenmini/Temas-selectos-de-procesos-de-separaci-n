**2.1 Bucle for: procesamiento de listas**

1. Crea una función que permita eliminar los valores nulos (None) de los valores numéricos contenidos en las siguientes listas.


image.png

***Solución:***

```python
#LISTAS PROPUESTAS
L1= [7.2, 7.8, 6.8, 8.0, None, '8.2', 5.6, 8.2, 7.7, 7.5, None, 5.8]
L2= ['6.8', None, 6.8, 6.1, 7.9, 9.4, None]
L3 = [8.5, '9.0', None, 7.7]


def quitar_None(lista):
  #En esta parte observamos que elimina los valores de None y convierte strings numéricos a "float"
  lista_limpia = []


  for elemento in lista:
    if elemento is not None:
      lista_limpia.append(float(elemento)) #no olvidar la conversión a número
  return lista_limpia
#Se aplica función a cada uno:
L1_limpia = quitar_None(L1)
L2_limpia = quitar_None(L2)
L3_limpia = quitar_None(L3)

print('L1 limpia:', L1_limpia)
print('L2 limpia:', L2_limpia)
print('L3 limpia:', L3_limpia)
```

```python
todos = [] #Esto lo hacemos para juntar todos los valores en una lista
for x in L1_limpia:
  todos.append(x)
for x in L2_limpia:
  todos.append(x)
for x in L3_limpia:
    todos.append(x)
print('Lista final:', todos)
```

**2.3 Bucle for: cadenas de texto**

5. Crea un programa que reciba una cadena de caracteres y determine si el texto
ingresado corresponde al nombre de un estado o a una capital de la República
Mexicana.

***Solución:***

```python
#Estados de la Republica Mexicana (con mayúscula inicial)
estados = [
    "Aguascalientes","Baja California","Baja California Sur","Campeche","Chiapas","Chihuahua",
    "Coahuila","Colima","Durango","Guanajuato","Guerrero","Hidalgo","Jalisco","Estado de México",
    "Michoacán","Morelos","Nayarit","Nuevo León","Oaxaca","Puebla","Querétaro","Quintana Roo",
    "San Luis Potosí","Sinaloa","Sonora","Tabasco","Tamaulipas","Tlaxcala","Veracruz","Yucatán",
    "Zacatecas","Ciudad de México"
]

#Capitales
capitales = [
    "Aguascalientes","Mexicali","La Paz","Campeche","Tuxtla Gutiérrez","Chihuahua",
    "Saltillo","Colima","Durango","Guanajuato","Chilpancingo","Pachuca",
    "Guadalajara","Toluca","Morelia","Cuernavaca","Tepic","Monterrey",
    "Oaxaca","Puebla","Querétaro","Chetumal","San Luis Potosí",
    "Culiacán","Hermosillo","Villahermosa","Ciudad Victoria","Tlaxcala",
    "Xalapa","Mérida","Zacatecas","CDMX"
]


texto = input("Escribe un estado o una capital de México: ")

#Comparación
texto_comp = texto.strip().lower()

es_estado = False
es_capital = False

for e in estados:
    if texto_comp == e.lower():
        es_estado = True

for c in capitales:
    if texto_comp == c.lower():
        es_capital = True

if es_estado and es_capital:
    print(f"'{texto}' Pertenece a un estado y también a una capital")
elif es_estado:
    print(f"'{texto}' Pertenece a un estado de la Republica Méxicana")
elif es_capital:
    print(f"'{texto}' Pertenece a una capital")
else:
    print(f"'{texto}' NO corresponde a un estado ni a una capital.")
```

**2.4 Bucle while: control por condición**

6. Utiliza un bucle while para solicitar números al usuario y agregarlos a una lista. El
programa deberá detenerse cuando la suma total de los elementos sea mayor
que 100. Al finalizar, muestra la lista completa y la suma total.

***Solución:***

```python
numeros = [] #Lista vacía para contener los números

suma = 0

# Bucle while: se repite mientras la suma sea menor o igual a 100
while suma <= 100:

    #Número al usuario
    num = float(input("Ingresa cualquier número: "))

    #Agregamos el número solicitado
    numeros.append(num)

    #Total
    suma += num

print("\nLista total de números:", numeros)
print("Suma final:", suma)
```

**2.5 Bucle while: menú de opciones**

7. Crea una lista vacía y desarrolla un programa que muestre un menú de opciones al
usuario utilizando un bucle while.

El menú deberá permitir realizar las siguientes acciones:


*   Agregar un número a la lista.
*   Eliminar el último número de la lista.


*   Mostrar el contenido actual de la lista.
*   Salir del programa.



El programa deberá repetirse mientras el usuario **no seleccione la opción de salir**. Para ello, utiliza una variable de control que permita terminar el bucle while
cuando el usuario así lo indique.

*Un posible esquema de inicio del menú es el siguiente:*


image.png

***Solución:***

```python
lista = [] #Lista vacía

opcion = "" #Control de menú


while opcion != "4": # Bucle while: se repite hasta que el usuario elija salir


    print("\nMenú:")
    print("1. Ingresa número")
    print("2. Eliminar último número")
    print("3. Mostrar lista")
    print("4. Salir")


    opcion = input("Seleccione una opción: ")


    if opcion == "1":
        num = float(input("Ingrese un número: "))
        lista.append(num)
        print("El número se agrego correctamente")

    elif opcion == "2":
        if len(lista) > 0:
            eliminado = lista.pop()
            print("Se borró el número:", eliminado)
        else:
            print("La lista está vacía, es imposible eliminar")

    elif opcion == "3":
        print("Actualización de la lista:", lista)


    elif opcion == "4":
        print("Saliendo del programa...")


    else:
        print("Intente más tarde")
```
