**1. Ley de los Gases Ideales**

La ecuación de los gases ideales relaciona la presión, el volumen, la cantidad de sustancia
y la temperatura mediante la expresión:
           image.png


Datos:

P= 1.2 atm

R= 0.0821 L * atm/ mol* K *(busqueda hecha)*

T= 273 K

V= 0.7 L

Realice una busqueda para obtener el valor de la constante de los gases (R) en las unidades
requeridas. Determina el valor de los moles n del sistema.


**Solución:**

*Lo primero que debemos hacer es despejar la ecuación (gases ideales) de la siguiente manera para determinar los moles del sistema dado:*

PV = nRT

*Pasamos dividiendo RT del otro lado:*

PV/ RT= n

*Por lo tanto queda de la siguiente manera: *

**n=PV/RT**

*Ahora solo sustituimos cada termino:*

```python
'Datos'
P = 1.2 #atm
R = 0.0821 #L*atm/mol* K # Este dato fue investigado para las unidades de los demás datos.
T = 273 #K
V = 0.7 #L

n = (P*V)/(R*T)
print(n) #En este caso pedimos que se nos imprima el resultado, pero no aparecen los datos, ni las cifras significativas que solicitamos para que se vea más formal.
```

Haremos un paso más, en este caso le pediremos a Phyton que nos de los decimales necesarios, de esta forma.

**6. Uso de condicionales: Determinación del régimen de flujo**

image.png

Re < 2300 → Flujo laminar

2300 ≤ Re ≤ 4000 → Flujo de transición

Re > 4000 → Flujo turbulento


**Solución**

```python
rho = float(input('Ingresa la densidad del fluido en kg/m3: ')) #float es utilizado para representar números reales, números decimales.
v = float(input('Ingresa la velocidad promedio del fluido en m/s: ')) #input se utiliza para pedirle datos al usuario y así poder hacer la operación matematica en este caso.
D = float(input('Ingresa el diámetro interno de la tubería en m: '))
mu = float(input('Ingresa la viscosidad dinámica del fluido en Pa/s: '))

Re = rho*v*D/mu

if Re < 2300:
  print('El flujo se mantiene en régimen laminar')
elif Re >= 2300 and Re <= 400:
  print('El sistema opera en flujo transitorio')
else:
  print('El flujo del sistema es turbulento')
```
